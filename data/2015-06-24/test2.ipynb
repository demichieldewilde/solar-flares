```python
import os
import numpy as np
from astropy.io import fits

# Define a function to merge FITS files
def merge_fits_files(fits_files, output_file):
    # Initialize an empty list to store the data
    data = []

    # Iterate over each FITS file
    for fits_file in fits_files:
        # Open the FITS file
        with fits.open(fits_file, memmap=True) as hdul:
            # Append the data to the list
            data.append(hdul[0].data)

    # Stack the data along the first axis
    data = np.stack(data, axis=0)

    # Create a new FITS file with the merged data
    hdu = fits.PrimaryHDU(data)
    hdu.writeto(output_file)

# Define the list of FITS files
fits_files = [f"fits/Ha/crisp_l2_20150624_134006_6563_r00{i+300}.fits" for i in range(512)]

# Define filename for the FITS file
filename_merged = 'crisp_134006_Ha_20150624.fits'

# Call the function to merge the FITS files and save the result as "merged.fits"
merge_fits_files(fits_files, filename_merged)
```

```python
from astropy.io import fits
import os

# Define a function to merge FITS files
def merge_fits_files(fits_files, output_file):
    # Open the first FITS file and write it to the output file
    with fits.open(fits_files[0], memmap=True) as hdul:
        hdul.writeto(output_file, overwrite=True)

    # Iterate over each remaining FITS file
    for fits_file in fits_files[1:]:
        try:
            # Try to open the FITS file with memory-mapping
            hdul = fits.open(fits_file, memmap=True)
        except ValueError:
            # If a ValueError occurs, open the FITS file without memory-mapping
            hdul = fits.open(fits_file, memmap=False)

        # Open the output file
        with fits.open(output_file, mode='update', memmap=False) as hdul_out:
            # Append the data from the current FITS file to the output file
            hdul_out.append(hdul[0])
            hdul_out.flush()  # changes are written back to disk

# Define the list of FITS files
fits_files = [f"file_{i}.fits" for i in range(513)]

# Call the function to merge the FITS files and save the result as "merged.fits"
merge_fits_files(fits_files, "merged.fits")
```
